```python
from IPython import get_ipython
from IPython.display import display
import pandas as pd
import numpy as np
import zipfile
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

# Extract the data
with zipfile.ZipFile('/content/weather_price_pre.zip', 'r') as ref:
  ref.extractall()

# Load the data
data = pd.read_csv('/content/seattle-weather.csv')

# Preprocess the data
x = data.drop('weather', axis=1)
y = data['weather']
date = x['date']
x = x.drop('date', axis=1)
scale = StandardScaler()
x = scale.fit_transform(x)
x_df = pd.DataFrame(x)
x_df.columns = scale.get_feature_names_out()
x = x_df
data = pd.concat([x, date], axis=1)
encoder_y = OneHotEncoder()
y = encoder_y.fit_transform(y.values.reshape(-1, 1)).toarray()
y_df = pd.DataFrame(y)
y_df.columns = encoder_y.get_feature_names_out(['weather'])
y = y_df
pca = PCA(n_components=5)
y = pca.fit_transform(y)
y = np.argmax(y, axis=1)
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, shuffle=True)

# Train and evaluate the model
def train(model, x_train, x_test, y_train, y_test):
  model.fit(x_train, y_train)
  score = model.score(x_test, y_test)
  print(score)

models = [
    LogisticRegression, DecisionTreeClassifier, RandomForestClassifier, SVC,
    GaussianNB, KNeighborsClassifier, XGBClassifier, LGBMClassifier,
    CatBoostClassifier
]

for i in models:
  print(i)
  train(i(), x_train, x_test, y_train, y_test)
```
