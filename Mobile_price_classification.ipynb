```python
import pandas as pd
import zipfile
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
with zipfile.ZipFile('/content/Mobile_price_classfaction.zip', 'r') as zip_ref:
    zip_ref.extractall()
```

```python
data = pd.read_csv('/content/train.csv')
```

```python
data['price_range'].unique()
```

```python
corr = data.corr()['price_range']
corr.plot(kind='bar', figsize=(10,5))
```

```python
plt.scatter(data['ram'], data['price_range'])
```

```python
x = data.drop('price_range', axis=1)
y = data['price_range']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
```

```python
y_train = y_train.values.reshape(-1,1)
y_test = y_test.values.reshape(-1,1)
```

```python
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(x_train, y_train)
knn.score(x_test, y_test)
```

```python
pre = knn.predict(x_test)
plt.scatter(x_test['ram'], y_test, c=pre)
plt.xlabel('ram')
plt.ylabel('price_range')
plt.title('KNN Prediction')
plt.show()
```
