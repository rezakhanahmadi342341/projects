```python
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
```

```python
data = pd.read_csv('/content/housing.csv')
```

```python
ocean_proximity = data['ocean_proximity']
data = data.drop(['ocean_proximity'], axis=1)
x = data.drop(['median_house_value'], axis=1)
y = data['median_house_value']
```

#**Imputer**

```python
emput = SimpleImputer(strategy='median')
data = emput.fit_transform(x)
empute_df = pd.DataFrame(data)
empute_df.columns = x.columns
x = empute_df
```

#**scale** **Data**

```python
scale = StandardScaler()
x = pd.DataFrame(scale.fit_transform(x), columns=x.columns)
x.shape
scale_y = StandardScaler()
y = pd.DataFrame(scale_y.fit_transform(y.values.reshape(-1,1)), columns=['median_house_value'])
y
```

# **OneHotEncoder**

```python
one_hot = OneHotEncoder(sparse_output=False)
one_hoted = one_hot.fit_transform(ocean_proximity.values.reshape(-1,1))
ocian_df = pd.DataFrame(one_hoted)
ocian_df.columns = ['<1H OCEAN', 'INLAND', 'ISLAND', 'NEAR BAY', 'NEAR OCEAN']
x = pd.concat([x, ocian_df], axis=1)
x
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
x = pd.DataFrame(pca.fit_transform(x))
x
```

#train a mudel

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=3, shuffle=True)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
score = model.score(x_test, y_test)
score
```

```python
pre = model.predict(x_test)
```

```python
mse = mean_squared_error(pre, y_test)
mse = np.sqrt(mse)
mse
```

```python
linear_models = [
    LinearRegression(),
    Ridge(),
    Lasso(),
    ElasticNet(),
    make_pipeline(PolynomialFeatures(degree=2), LinearRegression()),  # Polynomial Regression
    SVR()
]
```

```python
def models (models, x_train, y_train, x_test, y_test):
  for model in models:
    model.fit(x_train, y_train)
    score = model.score(x_test, y_test)
    print(f"{model} score: {score}")
    print('-' * 100)
```

```python
models(linear_models, x_train, y_train, x_test, y_test)
```
